```python
#  Prepare data for pie chart
a1 = df1['Fulfilment'].value_counts()
# Step 4: Plot the pie chart
fig, ax = plt.subplots()

ax.pie(a1, labels=a1.index, autopct='%1.1f%%', radius=0.7, wedgeprops=dict(width=0.6))
ax.set(aspect="equal")

plt.show()
```

```python
# Prepare data for scatter plot
x_data = df1['Category']  
y_data = df1['Size'] 

# Plot the scatter plot
plt.scatter(x_data, y_data)
plt.xlabel('Category ')  
plt.ylabel('Size')  
plt.title('Scatter Plot') 
plt.show()
```

```python
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df1, x='ship-state',hue="ship-state")
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

```python
# top_10_States 
top_10_state = df1['ship-state'].value_counts().head(10)
# Plot count of cities by state
plt.figure(figsize=(12, 6))
sns.countplot(data=df1[df1['ship-state'].isin(top_10_state.index)], x='ship-state',hue="ship-state")
plt.xlabel('ship-state')
plt.ylabel('count')
plt.title('Distribution of  State')
plt.xticks(rotation=45)
plt.show()
```

Note: From above Graph you can see that most of the buyers are Maharashtra state
